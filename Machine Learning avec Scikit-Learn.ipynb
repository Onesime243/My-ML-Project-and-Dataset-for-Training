# Listes vs. Tableaux en Python : Une Explication Détaillée

En Python, les listes et les tableaux sont deux structures de données utilisées pour stocker des collections d'éléments. Bien qu'ils puissent sembler similaires au premier abord, ils présentent des différences importantes en termes de performances, de flexibilité et d'utilisation.

**Les Listes**

- **Nature** : Les listes sont des objets mutables, c'est-à-dire qu'on peut modifier leur contenu après leur création. Elles peuvent stocker des éléments de différents types (entiers, chaînes de caractères, autres listes, etc. ).
- **Création**: On crée une liste en utilisant des crochets [ ].
- **Utilisation** : Elles sont très polyvalentes et sont utilisées pour une grande variété de tâches, telles que stocker des données hétérogènes, implémenter des piles, des fichiers d'attente, etc.
Exemple:

```python
ma_liste = [1, "hello", 3.14, [4, 5]]
```

**Les Tableaux (NumPy Arrays)**

- **Nature** : Les tableaux (implémentés par le module NumPy) sont des objets mutables optimisés pour les opérations numériques sur des données homogènes (généralement des nombres). Ils sont plus efficaces que les listes pour les calculs numériques.
- **Création** : On crée un tableau à l'aide de la fonction array()du module NumPy.
- **Utilisation** : Ils sont principalement utilisés pour des opérations mathématiques et scientifiques, comme la manipulation de matrices, le calcul vectoriel, etc.
Exemple:

```python
import numpy as np

mon_tableau = np.array([1, 2, 3, 4])
```

**Quand Utiliser Quoi ?**

- **Listes:**
  - Stockage de données hétérogènes
  - Structures de données complexes
  - Opérations générales sur les éléments
  
- **Tableaux:**
  - Les calculs numériques s'intensifient
  - Manipulation de matrices et de vecteurs
  - Opérations sur des données numériques homogènes

```python
# Avec une liste
ma_liste = [1, 2, 3, 4, 5]
moyenne = sum(ma_liste) / len(ma_liste)
print(moyenne)

# Avec un tableau NumPy
import numpy as np
mon_tableau = np.array([1, 2, 3, 4, 5])
moyenne = np.mean(mon_tableau)
print(moyenne)
```

**En résumé**:

Le choix entre une liste et un tableau dépend principalement de la nature des données que vous manipulez et des opérations que vous souhaitez effectuer. Si vous avez besoin d'une structure de données flexible pour stocker des éléments de différents types, une liste est plus adaptée. Si vous avez besoin d'effectuer des calculs numériques sur des données homogènes, un tableau NumPy sera plus efficace.

# Création de tableaux multidimensionnels en Python

Un tableau multidimensionnel est une structure de données qui permet de stocker des éléments organisés en plusieurs dimensions. En Python, nous utilisons souvent des listes imbriquées pour représenter ces tableaux.

```python
# Créer une matrice 3x3
matrice = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]

print(matrice)
```

**Commentaire**: Dans cet exemple, est une liste de listes. Chaque liste interne représente une ligne de la matrice. matrice

```python
# Accéder à l'élément en deuxième ligne, troisième colonne
element = matrice[1][2]
print(element)  # Affiche 6
```

**Créer un tableau de taille quelconque**

```python
# Créer un tableau de taille m x n rempli de zéros
def creer_tableau(m, n):
    tableau = [[0 for _ in range(n)] for _ in range(m)]
    return tableau

# Exemple d'utilisation
tableau = creer_tableau(3, 4)
print(tableau)
```

**Utiliser NumPy pour des opérations plus efficaces**

Pour des opérations numériques sur des tableaux multidimensionnels, la bibliothèque NumPy est très efficace. Elle fournit des objets ndarrayqui sont optimisés pour les calculs scientifiques.

Les vecteurs propres sont largement utilisés dans les bibliothèques d'apprentissage automatique. Intuitivement, étant donné une transformation linéaire représentée par une matrice A, les vecteurs propres sont des vecteurs qui, lorsque cette transformation est appliquée, ne changent que d'échelle (et non de direction). Plus formellement:

                                                    Av = λv
                                                    
où A est une matrice carrée, λ contient les valeurs propres et v contient les vecteurs propres. Dans l'ensemble d'outils d'algèbre linéaire de NumPy, eig nous permet de calculer les valeurs propres et les vecteurs propres de n'importe quelle matrice carrée.

```python
matrice = np.array([[1, -1, 3],
                   [1, 1, 6],
                   [3, 8, 9]]) #Creation d'une matrice de 3 lignes et 3 colonnes.

# Calculer les valeurs propres et les vecteurs propres
eigenvalues, eigenvectors = np.linalg.eig(matrice)
#Affiche les valeurs propres
eigenvalues
```

**1.15 Calcul des produits scalaires**

Le produit scalaire de deux vecteurs, a et b, est défini comme :


$$ \sum_{i=1}^n a_ib_i$$


où $a_i$ est le ième élément du vecteur a. Nous pouvons utiliser la classe dot de NumPy pour calculer le produit scalaire. Alternativement, dans Python 3.5+, nous pouvons utiliser le nouvel opérateur @ :

```python
# Création de deux vecteurs
vecteur_a = np.array([1,2,3])
vecteur_b = np.array([4,5,6])
# Calculate dot product
np.dot(vecteur_a, vecteur_b)
```

```python
# Calculate dot product
vecteur_a @ vecteur_b
```

# 1. Duo de Pandas & Numpy

NumPy (Numerical Python) et Pandas sont deux piliers fondamentaux en Python pour le traitement de données numériques. Bien qu'ils soient souvent utilisés ensemble, ils ont chacun des rôles spécifiques :

**1. NumPy : Les fondements**

- **Tableaux multidimensionnels** : NumPy est principalement conçu pour travailler avec des tableaux multidimensionnels (ndarray), qui sont des structures de données homogènes (tous les éléments doivent être du même type). Ces tableaux sont optimisés pour les calculs numériques, offrant une performance bien supérieure aux listes Python standard.
- **Opérations vectorielles et matricielles** : NumPy permet d'effectuer des opérations mathématiques complexes sur des tableaux entiers de manière très efficace, entraînant ainsi les boucles pour lentes.
- **Fonctions mathématiques** : Une vaste gamme de fonctions mathématiques est disponible dans NumPy, telles que le sinus, le cosinus, l'exponentielle, etc.

**Pourquoi NumPy est-il important ?**

- **Base pour d'autres bibliothèques** : NumPy sert de fondation pour de nombreuses autres bibliothèques de data science en Python, comme Pandas, SciPy, Matplotlib, etc.

- **Performance** : Les opérations sur les tableaux NumPy sont généralement beaucoup plus rapides que les opérations équivalentes en Python pur.

- **Simplicité** : NumPy offre une syntaxe concise et intuitive pour les opérations sur les tableaux.

**2. Pandas : Les données structurées**

- **DataFrames** : Pandas introduit la structure de données DataFrame, qui est essentiellement un tableau de données étiqueté, similaire à une feuille de calcul. Les DataFrames sont très flexibles et permettent de stocker des données de différents types (numériques, textes, dates, etc. ).

- **Indexation et sélection** : Pandas offre des outils puissants pour sélectionner et manipuler des données dans les DataFrames, tels que l'indexation par étiquette, la sélection par condition, etc.

- **Analyse de données** : Pandas fournit de nombreuses fonctions pour l'analyse de données, comme le calcul de statistiques descriptives, le regroupement, la fusion de DataFrames, etc.

**Pourquoi les pandas sont-ils importants ?**

- **Traitement de données réelles** : Pandas est idéal pour charger, nettoyer et manipuler des données réelles provenant de fichiers CSV, Excel, bases de données, etc.

- **Analyse exploratoire** : Pandas facilite grandement l'exploration de données, permettant de visualiser rapidement les principales caractéristiques d'un ensemble de données.

- **Préparation des données pour le machine learning** : Pandas est souvent utilisé pour préparer les données avant de les entraîner sur des modèles de machine learning.

**En résumé**

NumPy est la base pour les calculs numériques en Python, offrant des tableaux multidimensionnels optimisés et des opérations mathématiques efficaces.
Pandas construit sur NumPy pour fournir des structures de données plus riches (DataFrames) et des outils plus spécifiques pour l'analyse et la manipulation de données.

Ensemble, NumPy et Pandas forment un écosystème puissant pour la data science en Python, permettant aux chercheurs et aux data scientists de :
